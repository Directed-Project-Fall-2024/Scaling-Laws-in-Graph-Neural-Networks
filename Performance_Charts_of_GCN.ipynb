```python
import pandas as pd
import matplotlib.pyplot as plt

# Define your datasets
homophilic_datasets = ["Cora", "CiteSeer", "PubMed"]
heterophilic_datasets = ["Squirrel", "Chameleon", "Texas"]

def plot_multiple_datasets(datasets, title):
    plt.figure(figsize=(10, 6))

    for dataset_name in datasets:
        # Load dataset-specific results
        df = pd.read_csv(f"{dataset_name}_results.csv")

        # Plot the test accuracy (or validation accuracy) for comparison
        plt.plot(df["epoch"], df["test_acc"], label=f"{dataset_name} (Test Acc)")

    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

# Plot combined results for homophilic datasets
plot_multiple_datasets(homophilic_datasets, "GCN Accuracy Trends on Homophilic Datasets (Cora, CiteSeer, PubMed)")

# Plot combined results for heterophilic datasets
plot_multiple_datasets(heterophilic_datasets, "GCN Accuracy Trends on Heterophilic Datasets (Squirrel, Chameleon, Texas)")
```
